```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

random_file = "outputs/single-intersection/random_conn0_ep0.csv"
ppo_file = "outputs/single-intersection/ppo_conn0_ep0.csv"

# Read the CSV files into DataFrames
random_df = pd.read_csv(random_file)
ppo_df = pd.read_csv(ppo_file)

# Add identifiers
random_df['agent'] = 'Random'
random_df['episode'] = 0
ppo_df['agent'] = 'PPO'
ppo_df['episode'] = 1  # Set to 1 to differentiate

# Concatenate the DataFrames for plotting
all_data = pd.concat([random_df, ppo_df], ignore_index=True)

# Plot 1: Total Departed Vehicles Over Steps
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=all_data,
    x="step",
    y="system_total_departed",
    hue="agent"
)
plt.title("Total Departed Vehicles Over Time")
plt.xlabel("Simulation Step")
plt.ylabel("Departed Vehicles")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Mean Waiting Time Over Steps
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=all_data,
    x="step",
    y="system_mean_waiting_time",
    hue="agent"
)
plt.title("Mean Waiting Time Over Time")
plt.xlabel("Simulation Step")
plt.ylabel("Mean Waiting Time (s)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 3: Total Backlogged Vehicles Over Steps
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=all_data,
    x="step",
    y="system_total_backlogged",
    hue="agent"
)
plt.title("Backlogged Vehicles Over Time")
plt.xlabel("Simulation Step")
plt.ylabel("Vehicles Backlogged")
plt.grid(True)
plt.tight_layout()
plt.show()
```
